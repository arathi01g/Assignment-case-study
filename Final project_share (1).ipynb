# Observations
Accidents:
Accidents occured mostly on fridays and least on Sundays. Most of the accidents occured on 8AM and between 3PM to 6PM (Office and school hours) 85% of the severity is Slight injury

Driver:
35% of the accident occured to drivers in age band 18-30 and 33% for drivers in age band 31-50. 93% of accidents occured to male drivers. Accidents occured mostly to persons having Junior High school educational background 

Road:
Accidents occured mostly on lanes having two-way (divided with broken lines road marking). Road alignment were accident occured mostly was tangent road with flat terrain with Yshaped type of Junction.

```python
#Replacing NormalNormal with Normal
data['Fitness_of_casuality'].replace({"NormalNormal": "Normal"}, inplace=True)
```

```python
#Since Time is irrelavant, dropping the column
data.drop('Time', axis=1, inplace=True)
data.head()
```

```python
# Unknown,unknown and na is replaced with null
data.replace({"Unknown": np.nan, "unknown": np.nan, "na": np.nan}, inplace=True)
```

```python
#checking unique values after replacing unknown
columns = data.columns
for col in columns:
    print ('\nFrequency of Categories for varible %s'%col)
    print (data[col].value_counts())
```

```python
#check null
data.isna().sum()
```

```python
for col in data.drop('Accident_severity', axis=1):
    g = sns.FacetGrid(data, col='Accident_severity', size=6, aspect=1, sharey=True)
    g.map(sns.countplot, col, palette = 'Dark2')
    plt.show()
```

```python
for col in data.drop('Accident_severity', axis=1):
    data_grouped = data.groupby(['Accident_severity',col]).size().reset_index().pivot(columns='Accident_severity',index=col,values=0)
    data_grouped.plot(kind='bar', stacked=True)
```

```python
sns.heatmap(data.corr())
```

```python
## Filling missing values with Mode as all the columns having missing values are categorical columns
col = [x for x in data.isna().sum()[data.isna().sum() != 0].index.tolist()]
for i in col:
    mode = data[i].mode()[0]
    data[i].fillna(mode, inplace=True)
```

```python
data.isna().sum()
```

```python
sns.pairplot(data, hue='Accident_severity')
plt.show()
```

```python
# Encoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
for col in data.columns:
    data[col] = label_encoder.fit_transform(data[col])
```
